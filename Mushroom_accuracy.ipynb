#### class에서 edible = 0, poisonous =1 이므로 1일 땐 독버섯이구나

#### 범주형이므로 KNN모델을 사용해야겠다.

```python
# loc[] : 라벨값 기반의 2차원 인덱싱
# iloc[] : 순서를 나타내는 정수 기반의 2차원 인덱싱
x=df1.loc[:,"cap-shape":]
y=df1["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#data split
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#make the model
# model=KNeighborsClassifier(n_neighbors= k ).fit(x_train,y_train)
model=KNeighborsClassifier(n_neighbors=1).fit(x_train,y_train)
```

```python
#예측
pred=model.predict(x_test)
pred
```

```python
#정확도
(pred==y_test).sum()/len(pred)
```

#### 정확도가 1이 나왔다.

```python
df = pd.DataFrame({'y_test': y_test, 'prediction': pred,"diff":(y_test-pred)})
df.head()
```

```python
train_acc=[]
test_acc=[]

for k in range(1,30):
    #모델 만들고
    model=KNeighborsClassifier(n_neighbors=k).fit(x_train,y_train)
    #정확도 측정
    acc_tr=model.score(x_train,y_train)
    acc_te=model.score(x_test,y_test)
    
    train_acc.append(acc_tr)
    test_acc.append(acc_te)
    
    print("데이터가 인접한 {k}개의 다른 데이터를 가질 때".format(k=k)) 
    print("학습 정확도는 {a}, 테스트 정확도는 {z}이다. ".format(a=acc_tr,z=acc_te))
    print("\n")
```
