#### 2014~2017년 마약복용자가 종류별 범죄를 일으킨 건수

```python
install.packages("readxl")
library(readxl)

df1=read_excel("C:\\Users\\cloud\\Anaconda_src\\data\\마약.xlsx")

df3=read_excel("C:\\Users\\cloud\\Anaconda_src\\data\\ma4.xlsx")

df4=read_excel("C:\\Users\\cloud\\Anaconda_src\\data\\ma6.xlsx")
df5=read_excel("C:\\Users\\cloud\\Anaconda_src\\data\\ma9.xlsx")
df6=read_excel("C:\\Users\\cloud\\Anaconda_src\\data\\ma10.xlsx")
```

### 정수형을 범주형으로 변환 

```python
df3$year <- as.factor(df3$year)
df4$year <- as.factor(df4$year)
df5$year <- as.factor(df5$year)
df6$year <- as.factor(df6$year)
```

#### 유독 강력범죄, 지능범죄, 폭력범죄의 수치가 높았다.

```python
ggplot(df3, 
       aes(x=범죄대분류,y=cnt, fill=year)) +  #색채우기
  geom_bar(stat="identity",position="dodge") +
  scale_fill_brewer(palette="Pastel1")+ggtitle("마약자가 일으킨 범죄")
```

#### 대체적으로 줄어드는 수치지만 2014년 지능범죄의 수치는 매우 눂다

```python
#강력범죄
ggplot(df6, 
       aes(x=범죄중분류,y=cnt, fill=year)) +  #색채우기
  geom_bar(stat="identity",position="dodge") +
  scale_fill_brewer(palette="Pastel1")+geom_text(aes(label=cnt),position=position_dodge(width=1)) +ggtitle("마약자가 일으킨 강력범죄")


#지능범죄  
ggplot(df4, 
       aes(x=범죄중분류,y=cnt, fill=year)) +  #색채우기
  geom_bar(stat="identity",position="dodge") +
  scale_fill_brewer(palette="Pastel1")+geom_text(aes(label=cnt),position=position_dodge(width=1)) +ggtitle("마약자가 일으킨 지능범죄")


#폭력범죄
ggplot(df5, 
       aes(x=범죄중분류,y=cnt, fill=year)) +  #색채우기
  geom_bar(stat="identity",position="dodge") +
  scale_fill_brewer(palette="Pastel1")+geom_text(aes(label=cnt),position=position_dodge(width=1))+ggtitle("마약자가 일으킨 폭력범죄")
```
