```python
from icrawler.builtin import BingImageCrawler
```

```python
bing_crawler = BingImageCrawler(
    downloader_threads=4,
    storage={'root_dir': './상의'} # value에 경로 입력
)
```

```python
bing_crawler.crawl(keyword="top", filters=None, offset=0, max_num=1000) # keyword에 검색어, max_num = 최대 파일수
```
