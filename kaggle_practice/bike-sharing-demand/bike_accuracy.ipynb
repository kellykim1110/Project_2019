## 데이터 불러오기

### kaggle에 올라와 있는 bike-sharing-demand의 데이터 이용

https://www.kaggle.com/c/bike-sharing-demand

```python
import pandas as pd

train = pd.read_csv("./data/bike-sharing-demand/train.csv")
test = pd.read_csv("./data/bike-sharing-demand/test.csv")
train=train.drop(["casual","datetime","registered"],axis=1)
print(train.shape)
print(test.shape)
```

```python
print(train.info())     # 결측치나 데이터 개수 또는 자료형 확인
print(train.columns)    # 열 이름 확인
print(train.shape)      # 행 열 확인 
print(train.describe()) # 요약값 확인
print(train.head())     # 앞의 데이터 몇줄
print(train.tail())     # 뒤의 데이터 몇줄
```

```python
x = train.loc[:,"season":"windspeed"]
print(x.info())  
y = train['count']
```

## 데이터셋 나누기

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2,
                                                    random_state=42)
```

## 모델 만들어 학습 및 예측

#### 1. 모델 만들기
#### 2. 모델 학습  model.fit()
#### 3. 모델 예측 model.predict()
#### 4. 모델 평가

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(x_train,y_train)
pred=model.predict(x_test)
pred
```

```python
import pandas as pd
df=pd.DataFrame({"y_test" : y_test,"prediction" : pred, "diff" : (y_test - pred)})
df
```

```python
#MAE, MSE 가 낮을 수록 좋다.

MAE=sum(abs(df["diff"]))/len(df["diff"])
MSE=sum((df["diff"])**2)/len(df["diff"])

import numpy as np

RMSE=np.sqrt(MSE)

print(MAE)
print(MSE)
print(RMSE)
```

```python
acc_tr=model.score(x_train,y_train)
acc_test=model.score(x_test,y_test)
print("훈련 세트의 정확도 : {:.2f}".format(acc_tr))  
print("테스트 세트의 정확도 : {:.2f}".format(acc_test))  
```
