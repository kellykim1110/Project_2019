## 모델링을 해보자

```python
X=df1.loc[:,sale_col]
Y=df1.SalePrice
X_test=df2.loc[:,sale_col]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val=train_test_split(X,Y,random_state=0)
```

```python
#선형회귀
import numpy as np
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
#랜덤포레스트
from sklearn.ensemble import RandomForestRegressor
#그래디언트부스팅
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr=LinearRegression().fit(X_train,y_train)
pred1=lr.predict(X_val)

df1=pd.DataFrame({"y_test" : y_val,"prediction" : pred1, "diff" : (y_val - pred1)})
mse=sum((df1["diff"])**2)/len(df1["diff"])
rmse=np.sqrt(mse)
print("LinearRegression score : {:.3f}".format(rmse))
acc_tr1=lr.score(X_train,y_train)
acc_test1=lr.score(X_val,y_val)
print("훈련 세트의 정확도 : {:.2f}".format(acc_tr1))  
print("테스트 세트의 정확도 : {:.2f}".format(acc_test1))  
```

```python
print("정규화,확장 전 데이터 셋 : ", X.shape, Y.shape)
normalize_X = MinMaxScaler().fit_transform(X)# 입력 데이터 정규화
ex_X = PolynomialFeatures(degree=2, include_bias=False).fit_transform(normalize_X) # 데이터 feature
print("정규화, 추가 생성 : ", ex_X.shape, Y.shape)
```

```python
X_train, X_val, y_train, y_val=train_test_split(ex_X,Y,random_state=0)
lr1=LinearRegression().fit(X_train,y_train)
pred2=lr1.predict(X_val)
df2=pd.DataFrame({"y_test" : y_val,"prediction" : pred2, "diff" : (y_val - pred2)})
mse=sum((df2["diff"])**2)/len(df2["diff"])
rmse=np.sqrt(mse)
print("정규화 후, LinearRegression score : {:.3f}".format(rmse))
acc_tr2=lr1.score(X_train,y_train)
acc_test2=lr1.score(X_val,y_val)
print("훈련 세트의 정확도 : {:.2f}".format(acc_tr2))  
print("테스트 세트의 정확도 : {:.2f}".format(acc_test2))  
```

```python
#릿지회귀
#alpha=1
ridge1 = Ridge().fit(X_train, y_train)
#alpha=10
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
#alpha=0.1
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
#alpha=00.1
ridge001 = Ridge(alpha=0.01).fit(X_train, y_train)
print("if alpha=1")
print("훈련 세트 점수 : {:.2f}".format(ridge1.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge1.score(X_val, y_val)))
print("if alpha=10")
print("훈련 세트 점수 : {:.2f}".format(ridge10.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge10.score(X_val, y_val)))
print("if alpha=0.1")
print("훈련 세트 점수 : {:.2f}".format(ridge01.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge01.score(X_val, y_val)))
print("if alpha=0.01")
print("훈련 세트 점수 : {:.2f}".format(ridge001.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge001.score(X_val, y_val)))
```

```python
#라쏘회귀
#alpha=1
lasso1 = Lasso().fit(X_train, y_train)
#alpha=10
lasso10 = Lasso(alpha=10).fit(X_train, y_train)
#alpha=0.1
lasso01 = Lasso(alpha=0.1).fit(X_train, y_train)
#alpha=00.1
lasso001 = Lasso(alpha=0.01).fit(X_train, y_train)
print("if alpha=1")
print("훈련 세트 점수 : {:.2f}".format(lasso1.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso1.score(X_val, y_val)))
print("if alpha=10")
print("훈련 세트 점수 : {:.2f}".format(lasso10.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso10.score(X_val, y_val)))
print("if alpha=0.1")
print("훈련 세트 점수 : {:.2f}".format(lasso01.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso01.score(X_val, y_val)))
print("if alpha=0.01")
print("훈련 세트 점수 : {:.2f}".format(lasso001.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lasso001.score(X_val, y_val)))
```

```python
lasso01 = Lasso(alpha=0.1).fit(X_train, y_train)
pred=lasso01.predict(X_val)
df=pd.DataFrame({"y_test" : y_val,"prediction" : pred, "diff" : (y_val - pred)})
mse=sum((df["diff"])**2)/len(df["diff"])
rmse=np.sqrt(mse)
print("정규화 후, LinearRegression score : {:.3f}".format(rmse))
```
